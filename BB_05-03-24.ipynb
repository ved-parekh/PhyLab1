```python
# A to B
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.1, 2.1, 0.1)
y = np.array((0.1, 0.2, 0.302, 0.403, 0.504, 0.606, 0.707, 0.809, 0.911, 1.013, 1.115, 1.218, 1.320, 1.424, 1.532, 1.632, 1.735, 1.838, 1.946, 2.0))

results= np.polyfit(x,y,1)
m = results[0]
c = results [1]
y_model=m*x+c

plt.scatter(x,y, color= 'red', label='data points')
plt.plot(x,y_model, color='black', label='fit')
plt.xlabel('Voltage (v)')
plt.ylabel('Current (mA)')
plt.title('A to B')
plt.legend()
```

```python
print('M:', m)
```
