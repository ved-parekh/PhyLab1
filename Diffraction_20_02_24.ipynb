```python
# x is n*lambda, y is sin theta
import numpy as np
import scipy
import matplotlib.pyplot as plt
x = np.arange(1,7)*0.000555
y=np.array((0.0105,0.0151, 0.0213, 0.0262, 0.0328, 0.0382))
results= np.polyfit(x,y,1)
m= results[0]
c=results[1]
y_model=m*x+c
plt.scatter(x,y, color='red', label='Data Points')
plt.plot(x, y_model, color='black', label='fit')
plt.xlabel('1/d (cm^-1)')
plt.ylabel('sin theta ($\Theta$)')
r_err = np.array([0.001])

plt.text(0.0005,0.020, fr'$y = {np.round(m,2)}x + {np.round(c,2)}$')
plt.errorbar(x,y,yerr=r_err, linestyle='none', capsize=5, ecolor='blue', label='Errorbar')
plt.legend()
```

```python
#x=1/d, y = sin theta
import numpy as np
import matplotlib.pyplot as plt


x=np.array((100,200,300,600))
y=np.array((0.054, 0.108, 0.162, 0.331))
plt.scatter(x, y, color='red')

results = np.polyfit(x, y, 1)
m = results[0]
c = results[1]

y_model = m*x + c
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_model,color='black', label='Fit')

plt.text(100,0.25, fr'$y = {np.round(m,7)}x  {np.round(c,7)}$')
plt.xlabel('m $\lambda$ (mm)')
plt.ylabel('sin theta ($\Theta$)')
r_err = np.array([0.01])
plt.errorbar(x,y,yerr=r_err, linestyle='none', capsize=5, ecolor='blue', label='Errorbar')
plt.legend()
```
