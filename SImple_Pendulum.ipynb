```python
#Part C
import numpy as np
import matplotlib.pyplot as plt
x= np.array([10,20,30,40,50,60,70,80])
y= np.array([13.07,13.17,13.36,13.49,13.74,13.94, 14.12, 14.52])

results = np.polyfit(x,y,1)
m= results[0]
c= results[1] 
y_model=m*x+c

plt.scatter(x,y, color='red', s=20, label= 'data points') 
plt.plot(x,y_model, color= 'black', label='fit')
plt.xlabel(r'x ($\Theta$)')
plt.ylabel('y (T)')
plt.legend()
#plt.text(3,8, fr'$y={np.round(m,2)}x + {np.round(c,2)}$')
r_err = np.array([0.1])
plt.errorbar(x,y,yerr=r_err, linestyle='none', capsize=5, ecolor='blue', label='Errorbar')

#plt.text(3,8,fr'$y = {np.round(m,2)}x + {np.round(c,2)}$')
```

```python
#Part B
import numpy as np
import matplotlib.pyplot as plt
y=np.array([8.50, 9.87,10.86, 11.70, 12.54, 13.32, 14.24, 14.82, 15.52, 16.20])
x=np.array([17.41, 22.41, 27.41, 32.41, 37.41, 42.41, 47.41, 52.41, 57.41, 62.41]) # length

y2 = y**2

results = np.polyfit(x,y2,1)
m= results[0]
c= results[1] 

y_model= m*x + c

plt.scatter(x,y2, color='red', s=20, label= 'data points') 
plt.plot(x,y_model, color= 'black', label='fit')
plt.xlabel('x (Length)')
plt.ylabel('y(T^2)')
plt.legend()
r_err = np.array([7])
plt.errorbar(x,y2,yerr=r_err, linestyle='none', capsize=5, ecolor='blue', label='Errorbar')
```
