```python
# part A
import numpy as np
import matplotlib.pyplot as plt
x = np.array((1, 2, 3, 4, 5)) # K
y = np.array((140.491, 224.116, 323.924, 434.983, 497.408)) # deg

results = np.polyfit(x, y, 1)
m = results[0]
c = results[1]


y_model = m*x + c


plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_model,color='black', label='Fit')

plt.xlabel('K:order')
plt.ylabel('Deviation ($\degree$)')
plt.legend()
```

```python
x = np.array((0.653, 0.68, 0.752)) # clove, gly, water
y = np.array((0.648, 0.689, 0.794)) # in degree
results = np.polyfit(x, y, 1)
m = results[0]
c = results[1]

y_model = m*x + c

plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_model,color='black', label='Fit')

plt.xlabel('1/$\mu$')
plt.ylabel('cos($\phi$/6) ($\degree$)')
plt.legend()
```
