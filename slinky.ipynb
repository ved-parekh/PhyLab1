```python
# extension from 0 value
import numpy as np
import matplotlib.pyplot as plt
x = np.array((0.012, 0.0252, 0.027, 0.0319, 0.0432)) # kg
y = np.array((0.0975, 0.204, 0.2165, 0.2595, 0.3535)) #m

results = np.polyfit(x, y, 1)
m = results[0]
c = results[1]

y_model = m*x + c
print("m:", m)
print("M:", 9.8/m)

plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_model,color='black', label='Fit')

plt.xlabel('Mass (kg)')
plt.ylabel('Extension (m)')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 20) * 0.255/100
y = np.array((5.65, 12.1, 18.9, 25.1, 31.3, 36.8, 42.3, 47.9, 51.9, 56, 59.9, 63.4, 66.4, 69.1, 71.5, 73.3, 74.9, 76.1, 76.9, 77.3))/100

e = y-x

m = 177.72/1000
g = 9.8
k = 1.19
l = 5.1/100

# y_th = (m*g/k)*((x/l)-0.5*(x/l)**2)
y_th = (m * g / k) * ((x / l) - (0.5 * (x / l) ** 2))


plt.scatter(x, e, color='red', label='Data Points')
# plt.plot(x, y_th, color='black', label='Fit')
plt.plot(x, y_th, color='blue', label='Theoretical Fit')

plt.xlabel('x')
plt.ylabel('E(x)')
plt.legend()
```
